```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import pyautogui as auto
import time

auto.PAUSE =2.0
```

```python
driver = webdriver.Chrome(executable_path= "/usr/bin/chromedriver")
driver.get("https://yellowpages.webindia123.com/o-d/Karnataka/Bangalore/2/1/")

#driver.findElement("<span class=""jcl"" onclick=""closePopUp('best_deal_div');").click();
#driver=driver.findElement(By.xpath("//span[contains(@class,'jcl') and contains(text(), 'best_deal_div')")
#driver.click()
#new WebDriverWait(driver, 30).until(ExpectedConditions.visibilityOfElementLocated(By.xpath("//span[contains(@class,'middle') and contains(text(), '')]"))).click()
    
# <span class="middle">Next</span>
#<span class="jcl" onclick="closePopUp('best_deal_div');">X</span>

auto.click(579, 24, clicks= 2)

time_end = time.time()+ 60*1
time.sleep(5)
auto.moveTo(1694, 808, duration=0.5)

while time.time()< time_end:
    continue
    
res = driver.execute_script("return document.documentElement.outerHTML")
soup = bs(res,'lxml')

find= soup.find_all('a', href= True)

url =[]
for i in find:
    if '/Bangalore/' in i.get('href'):
        url.append(i.get('href'))
    
```

```python
base ='https://yellowpages.webindia123.com'

doctor =[]

for i in range(2,len(url)):
    req = requests.get(base+url[i])
    soup = bs(req.text,"lxml")
    title = soup.find_all('h1')
    print(len(title))
    if len(title)!=0:
        title = title[0].text
    
    name = soup.find_all('h4')
    if len(name)!=0:
        name = name[0].text
    
    loc = soup.find_all('div' ,class_='loc')
    if len(loc)!=0:
        loc = loc[0].text
    
    temp =[]
    
    temp.append(name)
    temp.append(loc)
    temp.append(title)
    
    
    #temp.append(date[0])
    
    doctor.append(temp)
```

```python
doc = pd.DataFrame(doctor, columns =['Name','Location','Speciality'])
```
